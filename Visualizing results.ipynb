```python
df1 = pd.read_csv(r'/Users/mac/Desktop/MrFrank/data3Results(full labels).csv')
df1.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the number of charts per batch and the number of metrics
num_charts = 6
metrics = df1.columns[1:]  # Assuming the first column is the 'Model' name
num_rows = 3
num_cols = 2

# Get the total number of models
total_models = len(df1)

# Iterate over the DataFrame in steps of `num_charts`
for start in range(0, total_models, num_charts): 
    # Create a new figure
    plt.figure(figsize=(14, 15))
    
    # Iterate over the models in the current batch
    for i in range(num_charts):
        if start + i < total_models:
            model_name = df1.iloc[start + i, 0]  # Get model name
            values = df1.iloc[start + i, 1:]  # Get the corresponding values
            
            # Debugging: print the shapes of metrics and values
            print(f"Model: {model_name}, Metrics Shape: {metrics.shape}, Values Shape: {values.shape}")
            
            # Check if the length of `metrics` matches the length of `values`
            if len(metrics) != len(values):
                print(f"Skipping model {model_name} due to length mismatch.")
                continue
            
            plt.subplot(num_rows, num_cols, i + 1)
            bars = plt.bar(metrics, values, color=sns.color_palette("husl", num_charts)[i])
            
            # Add labels on each bar
            for bar in bars:
                yval = bar.get_height()
                plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.5, round(yval, 2), ha='center', va='bottom')

            plt.title(model_name)
            plt.xlabel('Metric')
            plt.ylabel('Score')
            plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
    
    # Adjust layout to prevent overlap
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```

```python
# Extract the unique models (e.g., Random Forest, XGBoost, etc.)
models = df1['Data balancing technique/model'].str.split().str[-1].unique()

# Set up the plotting grid
num_models = len(models)
fig, axes = plt.subplots(nrows=num_models, ncols=1, figsize=(10, 5 * num_models))

# Loop through each model and plot
for i, model in enumerate(models):
    # Filter the DataFrame for the current model
    df_filtered = df1[df1['Data balancing technique/model'].str.contains(model)]
    
    # Melt the DataFrame for easier plotting
    df_melted = df_filtered.melt(id_vars=['Data balancing technique/model'], 
                                 var_name='Metric', 
                                 value_name='Score')
    
    # Plot the data
    sns.barplot(x='Metric', y='Score', hue='Data balancing technique/model', data=df_melted, ax=axes[i], palette="turbo")
    
    # Set plot title and adjust layout
    axes[i].set_title(f'{model} Performance Across Different Data Balancing Techniques', fontsize=10)
    axes[i].set_xlabel('Evaluation Metric')
    axes[i].set_ylabel('Score')
    axes[i].legend(loc='best', title='Data Balancing Technique/Model')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
